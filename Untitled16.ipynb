```python
# 1. Necessary Libraries Import Karna
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# 2. Dummy Data Create Karna (Taaki aapka code bina kisi external file ke chal sake)
# Agar aapke paas apni CSV file hai, toh aap use kar sakte hain: data = pd.read_csv('filename.csv')
data_dict = {
    'Study_Week': [2, 4, 5, 6, 7, 8, 9, 10],
    'Marks': [40, 55, 60, 68, 72, 80, 88, 92]
}
data = pd.DataFrame(data_dict)

# 3. Features (X) aur Target (y) Split Karna
# Linear Regression ke liye X ko hamesha 2D array ya DataFrame hona chahiye
X = data[['Study_Week']]  
y = data['Marks']

# 4. Model Train (Fit) Karna
model = LinearRegression()
model.fit(X, y)

# 5. Predictions Nikalna
predicted_score = model.predict(X)

# 6. Model Performance Evaluation
r2 = r2_score(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

print(f"R2 Score: {round(r2, 2)}")
print(f"MSE: {round(mse, 2)}")
print(f"RMSE: {round(rmse, 2)}")
print("-" * 40)

# 7. Histogram Plot Karna (Aapke notes ke mutabiq)
plt.figure(figsize=(6, 4))
plt.hist(data['Marks'], bins=5, color='green', edgecolor='black') # Bins ko thoda kam kiya taaki dummy data par sahi dikhe
plt.title("Distribution of Marks")
plt.grid(True)
plt.show()

# 8. Scatter Plot aur Regression Line Plot Karna
plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, predicted_score, color='red', label='Regression Line')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours per Week")
plt.ylabel("Marks Obtained")
plt.legend()
plt.grid(True)
plt.show()

# 9. Naye Data Par Predict Karna (E.g., 9 Hours study ke liye)
new_hour = 9
# Model predict hamesha 2D array expect karta hai: [[value]]
predicted_score_new = model.predict([[new_hour]]) 

print(f"Predicted final score for {new_hour} hours is: {round(predicted_score_new[0], 2)}")
```
